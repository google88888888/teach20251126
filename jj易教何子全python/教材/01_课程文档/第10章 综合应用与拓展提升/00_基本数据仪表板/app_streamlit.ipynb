```python
# 获取数据
df = px.data.gapminder()
# 将列标题翻译为中文
df.columns = ['国家', '大洲', '年份', '预期寿命', '人口', '人均GDP', 'ISO代码', 'ISO编号']
df
```

```python
# 按国家汇总GDP并降序排序取前20名
df_sorted = df.groupby('国家')['人均GDP'].mean().sort_values(ascending=False).head(20).iloc[::-1]

# 绘制横向条形图
fig = px.bar(df_sorted,
             x=df_sorted.values,  # x轴数据(人均GDP)
             y=df_sorted.index,   # y轴数据(国家)
             orientation='h',      # 设置为横向
             title='各国平均人均GDP前20名对比',
             height=800, width=500,
             labels={'x': '平均人均GDP (美元)', 'y': '国家'})

# 翻转图表使x轴在上方
fig.update_layout(
    autosize=False,
    xaxis={'side': 'top'}
)

fig.show()
```

```python
# 按大洲和年份计算平均预期寿命
df_life = df.groupby(['大洲'])['预期寿命'].mean().reset_index()

# 绘制树状图
fig = px.treemap(df_life,
                 path=['大洲'],  # 定义层级路径
                 values='预期寿命',      # 数值大小决定矩形面积
                 title='各大洲不同年份的平均预期寿命',
                 height=600)

# 更新布局
fig.update_layout(
    title_x=0.5,  # 标题居中
    margin=dict(t=50, l=25, r=25, b=25)  # 调整边距
)

fig.show()
```

```python
# 按年份和大洲计算平均预期寿命
df_life_by_year = df.groupby(['国家', '大洲'])['预期寿命'].mean().reset_index()

# 绘制旭日图
fig = px.sunburst(df_life_by_year,
                  path=['大洲', '国家'],  # 定义层级路径
                  values='预期寿命',      # 数值大小决定扇形面积
                  title='各大洲不同国家的平均预期寿命',
                  height=700)

# 更新布局
fig.update_layout(
    title_x=0.5,  # 标题居中
    margin=dict(t=50, l=25, r=25, b=25)  # 调整边距
)

fig.show()
```
